Now we have list of raw features, we can try to count them

```python
features = ['ssn','zip5','dob','address','firstname_lastname',
           'homephone_address','homephone_zip5','address_zip5',
           'ssn_address','ssn_dob','ssn_zip5','ssn_homephone',
           'firstname_lastname_dob','firstname_lastname_ssn',
           'firstname_lastname_homephone','firstname_lastname_address',
           'firstname_lastname_zip5','address_zip5_homephone',
           'firstname_lastname_ssn_dob']
```

```python
len(features)
```

## this is a sample trial

```python
start = time.time()
##############
lag = 3 #this is the number of lags
#############
dic = {}
for i in range(len(dates)):
    for f in features:
        if i == 0:   #here we intialize the dictionary
            temp = subset.query('date_index <= 0')      
            dic['counter_'+f] = Counter(temp[f])
            dic['temp_count_'+f] = subset.query('date_index== @i')[f].apply(lambda x: dic['counter_'+f][x])
            dic[f+'_'+str(lag)] = dic['temp_count_'+f]
        elif 1 <= i <= lag-2:
            temp = subset.query('date_index <= @i')      
            dic['counter_'+f] = Counter(temp[f])
            dic['temp_count_'+f] = subset.query('date_index== @i')[f].apply(lambda x: dic['counter_'+f][x])
            dic[f+'_'+str(lag)] = np.hstack((dic[f+'_'+str(lag)],dic['temp_count_'+f]))
        else:
            temp = subset.query('@i-@lag+1 <= date_index <= @i') 
            dic['counter_'+f] = Counter(temp[f])
            dic['temp_count_'+f] = subset.query('date_index== @i')[f].apply(lambda x: dic['counter_'+f][x])
            dic[f+'_'+str(lag)] = np.hstack((dic[f+'_'+str(lag)],dic['temp_count_'+f]))

end = time.time()
print('time elapsed: {}'.format(str(end-start)))
```

## Now we know the trial works, we can scale it up

```python
lag_list = [1,3,7,15,30]
```

```python
dic = {}
for lag in lag_list:
    start = time.time()  
    for i in range(len(dates)):
        for f in features:
            if i == 0:   #here we intialize the dictionary
                temp = subset.query('date_index <= 0')      
                dic['counter_'+f] = Counter(temp[f]) ## count the factors and count the number
                dic['temp_count_'+f] = subset.query('date_index== @i')[f].apply(lambda x: dic['counter_'+f][x])
                dic[f+'_'+str(lag)] = dic['temp_count_'+f]
            elif 1 <= i <= lag-2:
                temp = subset.query('date_index <= @i')      
                dic['counter_'+f] = Counter(temp[f])
                dic['temp_count_'+f] = subset.query('date_index== @i')[f].apply(lambda x: dic['counter_'+f][x])
                dic[f+'_'+str(lag)] = np.hstack((dic[f+'_'+str(lag)],dic['temp_count_'+f]))
            else:
                temp = subset.query('@i-@lag+1 <= date_index <= @i') 
                dic['counter_'+f] = Counter(temp[f])
                dic['temp_count_'+f] = subset.query('date_index== @i')[f].apply(lambda x: dic['counter_'+f][x])
                dic[f+'_'+str(lag)] = np.hstack((dic[f+'_'+str(lag)],dic['temp_count_'+f]))

    end = time.time()
    print('lag{}, time elapsed: {}'.format(str(lag),str(end-start)))
```

### so total time elapsed was about 5.75 minutes on my laptop...
