```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv('train_test.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
labels = data['fraud']
features = data.iloc[:,:25]
features.head()
##split the label and the feature
```

```python
from sklearn.cross_validation import train_test_split
x_train, x_test,y_train,y_test = train_test_split(features, labels,test_size = 0.2, random_state = 0)
print(sum(y_train)) #fraud rate of training data
print(sum(y_test)) #fraud rate of testing data
```

### Random Forest Model

```python
rf_clf = RandomForestClassifier(max_depth=15,n_estimators=50, min_samples_leaf=50,
                                  min_samples_split=100, random_state=0)
rf_clf.fit(x_train,y_train)
imp_feat_rf = pd.Series(rf_clf.feature_importances_, index=x_train.columns).sort_values(ascending=False)

print(len(imp_feat_rf))
imp_feat_rf
```

Above is the feature importance extent by descending order

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score

def test_result(clf):
    rf_clf.fit(x_train, y_train)
    prediction_train = rf_clf.predict(x_train)
    matrix = confusion_matrix(y_train, prediction_train)
    train_acc = rf_clf.score(x_train, y_train)
    train_auc = roc_auc_score(y_train, prediction_train)
    train_fdr = matrix[1,1]/sum(y_train)

    prediction_test = rf_clf.predict(x_test)
    matrix = confusion_matrix(y_test, prediction_test)
    test_acc = rf_clf.score(x_test, y_test)
    test_auc = roc_auc_score(y_test, prediction_test)
    test_fdr = matrix[1,1]/sum(y_test)

    comparison_df = pd.DataFrame({ "Accuracy": [train_acc, test_acc],
                                  "AUC score": [train_auc, test_auc],
                                  "FDR": [train_fdr, test_fdr]},
                                 index = ["Training", "Testing"])
    return comparison_df
    

test_result(rf_clf)
```

### Adjusting 

```python
# Define a new Random Forest Classifier
from sklearn.grid_search import GridSearchCV
select_rf_clf = RandomForestClassifier(random_state=10)

param_grid = {
    
            'n_estimators': [50, 80, 100],
            'max_depth': [5,10, 15]
}

# we will 10-fold cross-validation
grid_clf = GridSearchCV(select_rf_clf,param_grid,cv=10)
grid_clf.fit(x_train,y_train)

# Take the best model
best_rf_clf = grid_clf.best_estimator_
predicted_proba = best_rf_clf.predict_proba(x_test)
predicted_proba
```
